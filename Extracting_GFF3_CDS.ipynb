```python
i=1
i2 = 1
#Don't worry about the i and i2. Kelsey put that in as a check to make sure the file was being processed. It has to do with the calculation below that shows how many megabytes have been processed
gff3_file = 'Elegans.gff3'
output_CDS = 'Elegans_CDS.gff3'
#input file is the GFF3 of the relevant C elegans release and the output file is where the filtered info will go 
#first part of the code here prints how many megabytes are processed from the original file 
with open(gff3_file,'r') as infile, open(output_CDS,'w') as outfile:
    while True:
        if i/(2**20) >= i2:
            print('Megabytes processed: %f' % (i/(2**20)))
            i2 += 1
        l = infile.readline()
        if len(l) == 0:
            break
#After reading the file in you want to trim the white space after each row and you want to create a list of strings for the data in each column. That is what l.strip and split do
        ltrim = l.strip()
        lcols = ltrim.split()
#To keep info at the top of GFF file you want to write the lines starting with #onto the output and then you want to write only those lines which have CDS in col 2 and WormBase in col 1
        if lcols[0][0] == '#':
            outfile.write(l)
        elif lcols[2] == 'CDS' and lcols[1] == 'WormBase':
            outfile.write(l)
        i+=len(l)
```
